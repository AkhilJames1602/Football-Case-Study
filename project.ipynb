### Data Processing

### Load the necessary dataset / Importing the required dataset

```python
# 01. Appearance Data
appearance = pd.read_excel('appearances.xlsx')
appearance
```

```python
appearance.isnull().sum()
```

```python
# 02. Game_event Data
game_event = pd.read_excel('game_events.xlsx')
game_event
```

```python
game_event.isnull().sum()
```

```python
game_event=game_event.drop(['player_in_id','player_assist_id'],axis=1) # this two column removed
```

```python
game_event["description"].mode()
```

```python
game_event["description"] = game_event["description"].fillna(", Tactical")
```

```python
game_event.isnull().sum()
```

```python
# 03. Game_Lineup
game_Lineup = pd.read_excel('game_lineups.xlsx')
game_Lineup
```

```python
game_Lineup.isnull().sum()
```

```python
# 04. Games Data
games = pd.read_excel('games.xlsx')
games
```

```python
games.isnull().sum()
```

```python
games=games.drop(['home_club_formation','away_club_formation'],axis=1)
```

```python
games.info()
```

```python
x = games["home_club_position"].mean()
games["home_club_position"] = games["home_club_position"].fillna(x)
```

```python
x = games["away_club_position"].mean()
games["away_club_position"] = games["away_club_position"].fillna(x)
```

```python
games["home_club_manager_name"].mode()
```

```python
games["home_club_manager_name"] = games["home_club_manager_name"].fillna("Mark Hughes")
```

```python
games["away_club_manager_name"].mode()
```

```python
games["away_club_manager_name"] = games["away_club_manager_name"].fillna("Mark Hughes")
```

```python
x = games["attendance"].mean()
games["attendance"] = games["attendance"].fillna(x)
```

```python
games["referee"].mode()
```

```python
games["referee"] = games["referee"].fillna("Dr. Felix Brych")
```

```python
games["home_club_name"].mode()
```

```python
games["home_club_name"] = games["home_club_name"].fillna("Borussia Verein für Leibesübung 1900 e.V.")
```

```python
games["away_club_name"].mode()
```

```python
games["away_club_name"] = games["away_club_name"].fillna("Borussia Dortmund")
```

```python
games.isnull().sum()
```

```python
# 05. Players Data
players = pd.read_excel('players.xlsx')
players
```

```python
players.isnull().sum()
```

```python
players.info()
```

```python
players["foot"].mode()
```

```python
players["foot"] = players["foot"].fillna("right")
```

```python
x = players["height_in_cm"].mean()
players["height_in_cm"] = players["height_in_cm"].fillna(x)
```

```python
x = players["market_value_in_eur"].mean()
players["market_value_in_eur"] = players["market_value_in_eur"].fillna(x)
```

```python
x = players["highest_market_value_in_eur"].mean()
players["highest_market_value_in_eur"] = players["highest_market_value_in_eur"].fillna(x)
```

```python
players["contract_expiration_date"].mode()
```

```python
players["contract_expiration_date"] = players["contract_expiration_date"].fillna("2025-06-30")
```

```python
players["agent_name"].mode()
```

```python
players["agent_name"] = players["agent_name"].fillna("Wasserman")
```

```python
players.isnull().sum()
```

### Merging the Data Files 

```python
df = pd.merge(appearance,players, on='player_id', how = 'left')
df
```

```python
df1 = pd.merge(df,game_event,on='game_id', how = 'left')
df1
```

```python
df2 = pd.merge(df1,game_Lineup, on='game_id', how = 'left')
df2
```

```python
df3 = pd.merge(df2,games, on='game_id', how = 'left')
df3
```

```python
df3.isnull().sum()
```

```python
df3=df3.drop(['competition_id_y','date_y', 'minute', 'player_id_y',
    'game_lineups_id', 'type_y', 'number', 'player_id',
       'player_name_y', 'team_captain', 'position_y'],axis=1)
```

```python
df3["description"].mode()
```

```python
df3["description"] = df3["description"].fillna(", Tactical")
```

```python
df3["type_x"].mode()
```

```python
df3["type_x"] = df3["type_x"].fillna("Substitutions")
```

```python
df3.isnull().sum()
```
